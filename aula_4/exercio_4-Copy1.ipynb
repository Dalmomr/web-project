```c++
#construindo a Tree
```

```c++
t = TTree("t", "t")
px = array('d', [0])
py = array('d', [0])
t.Branch("x", px, "x/D")
t.Branch("y", py, "y/D")
```

```c++
#anexando valores na tree
```

```c++
for i in range(100):
    px[0] = gRandom.Gaus(0, 3)
    py[0] = gRandom.Uniform() * 30 - 15
    t.Fill()
```

```c++
#importanto os dados para rootdatset
```

```c++
x = RooRealVar("x", "x", -3, 3)
y = RooRealVar("y", "y", -10, 10)
ds = RooDataSet("ds", "ds", RooArgSet(x, y),RooFit.Import(t))
ds.Print()
     
```

```c++
frame = y.frame(RooFit.Title("Importanto dados "))
ds.plotOn(frame)
```

```c++
c1 =  TCanvas()
ds.plotOn(frame)
frame.Draw()
c1.Draw()
```
